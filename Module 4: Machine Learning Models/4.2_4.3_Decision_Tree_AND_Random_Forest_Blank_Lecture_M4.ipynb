**Note:**

The confusion matrix shows:

*   True Positive (TP): 109
*   True Negative (TN):  59
*   False Positive (FP):  3
*   False Negative (FN):  0

We can find the accuracy by doing:

**Accuracy = (TP + TN)/(TP + TN + FP + FN)**

**Accuracy = (109+59)/(109+59+3+0)= 0.9825** 

This means that the accuracy using the confusion matrix is 98.25%.

We can also calculate the following:

**Precision:**

Precision = TP/(TP+FP) = 109/(109+3) = 0.9732 -> 97.32%

Precision calculates the model's ability to classify positive values correctly.

**Recall:**

Recall = TP/(TP+FN) = 109/(109+0) = 1.0 -> 100%

Recall calculates the model's ability to predict positive values.

**F1-Score:**

F1-Score = (2*Precision*Recall)/(Precision+Recall)

F1-Score = (2*0.9732*1.0)/(0.9732+1.0) = 0.9864 -> 98.64%
